## GÖRÜNTÜDEKİ KÖŞELERİ TESPİT ETME

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread(r"materyal\figureCorner.png")
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

corners = cv2.goodFeaturesToTrack(imgGray, 28, 0.01, 10) # 27 tane köşe bulacak şekilde ayarlandı
corners = np.int0(corners) # corners'ı int yapar

for corner in corners:
    x, y = corner.ravel() # ravel() fonksiyonu, bir diziye uygulandığında, dizi elemanlarını düzleştirir.
    cv2.circle(img, (x, y), 3, 255, -1) # 3 çapında beyaz bir daire çizilir

plt.imshow(img)
plt.show()
```
