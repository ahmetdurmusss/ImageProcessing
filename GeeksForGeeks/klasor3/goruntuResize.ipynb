## GÖRÜNTÜ YENİDEN BOYUTLANDIRMA

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"materyal\tomato.png")

half = cv2.resize(img, (0, 0), fx = 0.5, fy = 0.5) # parametreler: resim, boyutlar, x ve y ekseninde ne kadar büyütüleceği
bigger = cv2.resize(img, (1050, 1610)) # parametreler: resim, boyutlar

stretch_near = cv2.resize(img, (780, 540), interpolation = cv2.INTER_NEAREST) # parametreler: resim, boyutlar, interpolasyon

Titles = ["Original", "Half", "Bigger", "Interpolation Nearest"]
images = [img, half, bigger, stretch_near]
count = 4

for i in range(count):
    plt.subplot(2, 2, i + 1)
    plt.title(Titles[i])
    plt.imshow(images[i])

plt.show()
```
