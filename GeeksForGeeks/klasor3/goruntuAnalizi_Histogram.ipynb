### HİSTOGRAM İLE GÖRÜNTÜ ANALİZİ

```python
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread(r"materyal\flowers.png")
cv2.imshow("Flowers",img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
histr = cv2.calcHist([img],[0],None,[256],[0,256])
plt.plot(histr)
plt.show()
```

```python
plt.hist(img.ravel(),256,[0,256])
plt.show()
```
